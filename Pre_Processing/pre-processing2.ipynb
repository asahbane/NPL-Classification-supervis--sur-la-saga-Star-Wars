# Resampling
Deux type : le oversampling et le undersampling :

##### Application de la régression logistique avec la technique du "Oversampling"
cette méthode consiste à ajouter des copies de la classe qui est peu existante dans notre jeu de données.

```python
from sklearn.utils import resample
```

```python
G = X5.copy()
G = prep_plus_frequents_D7(G)
G
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(G.drop('gender', axis = 1), G['gender'])

M = pd.concat([X_train, Y_train], axis=1)

male = M[M.gender==0]
female = M[M.gender==1]

female_upsampled = resample(female,
                          replace=True, # sample with replacement
                          n_samples=len(male), # match number in majority class
                          random_state=27)

upsampled = pd.concat([male, female_upsampled])
```

# Weighted Logistic Regression
Cette methode consiste à associer des poids à la chaque classe afin de ratrapper un peu le manque d'une classe

```python
x1 = Counter(G['gender'])[0]/(Counter(G['gender'])[1]+Counter(G['gender'])[0])
x2 = Counter(G['gender'])[1]/(Counter(G['gender'])[1]+Counter(G['gender'])[0])
```

```python
print('Le pourcentage des hommes dans nos données est :', x1)
print('Le pourcentage des femmes dans nos données est :', x2)
```

```python
## dictionnaire des poids
w = {0 : Counter(G['gender'])[1]/(Counter(G['gender'])[1]+Counter(G['gender'])[0]),
    1 : Counter(G['gender'])[0]/(Counter(G['gender'])[1]+Counter(G['gender'])[0])}
```
